# 04 - Sales Rep / Team Performance Analysis

Key question (from segmentation findings): is Poland underserved due to lower rep coverage,
or is it genuinely a smaller market with similar resourcing? Also: which teams/reps cover
which customer segments (Key Accounts, Rising Stars, etc).

```python
import sys
sys.path.append("../src")

import pandas as pd
import matplotlib.pyplot as plt
from eda_utils import load_clean_data, save_fig
from segmentation import filter_valid_sales, add_order_type
from rep_performance import (
    rep_kpis, yoy_growth_by_rep, country_resource_allocation,
    rep_country_breakdown, plot_rep_kpis, plot_country_allocation,
    plot_rep_country_heatmap, team_segment_alignment
)
```
