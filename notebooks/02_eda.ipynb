# 02 - Exploratory Data Analysis

Explore sales trends, country/channel/product breakdowns, sales team performance, and geographic distribution.

```python
import sys
sys.path.append("../src")

import pandas as pd
import matplotlib.pyplot as plt
from eda_utils import (
    load_clean_data, save_fig, overall_summary,
    sales_trend_over_time, yoy_growth,
    sales_by_country, sales_by_channel,
    sales_by_product_class, top_products,
    sales_by_team, sales_by_rep, sales_by_manager, team_yearly_trend,
    geo_scatter
)
```

```python
df = load_clean_data("../data/processed/cleaned_data.csv")
df.head()
```

## Overall Summary

```python
summary = overall_summary(df)
for k, v in summary.items():
    print(f"{k}: {v}")
```

## 1. Sales Trend Over Time

```python
fig, trend = sales_trend_over_time(df, freq="M")
save_fig(fig, "01_monthly_sales_trend", path="../reports/figures")
plt.show()
```

```python
growth = yoy_growth(df)
growth
```
